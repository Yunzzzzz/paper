 # 2019、2020、2021年本科生毕业论文数据获取

```python
import requests
import re
import csv
```

## 北京师范大学地理科学学部2019届本科毕业论文答辩安排

```python
url = 'https://geo.bnu.edu.cn/tzgg/74359.html'
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.80 Safari/537.36 Edg/98.0.1108.43"
}
resp = requests.get(url,headers =headers)
resp.encoding= 'utf-8'
s = resp.text
```

```python
obj = re.compile(r'<tr height="33" style="height:27px;">.*?<td align="center" height="33" style="height:27px;" x:num=".*?">.*?</td>.*?<td align="center" x:str="">(?P<答辩人>.*?)'
                 r'</td>.*?<td x:str="">(?P<答辩人专业>.*?)</td>.*?<td x:str="">(?P<论文题目>.*?)'
                 r'</td>.*?<td align="center" style="width:146px;" x:str="">(?P<指导教师>.*?)'
                 r'</td>.*?</tr>',re.S)
result = obj.finditer(s)

f = open("北京师范大学地理科学学部2019届本科毕业论文答辩安排.csv",mode="w",newline='')#newline=''解决空行问题
csvwriter =csv.writer(f)

for it in result:
#     print(it.group("答辩人"))
#     print(it.group("答辩人专业"))
#     print(it.group("论文题目"))
#     print(it.group("指导教师"))
    dic = it.groupdict()
    csvwriter.writerow(dic.values())
    
f.close()
print("over!")
```

## 北京师范大学地理科学学部2019届本科毕业论文答辩安排

```python
url = 'https://geo.bnu.edu.cn/tzgg/113876.html'
headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.80 Safari/537.36 Edg/98.0.1108.43"
}
resp = requests.get(url,headers =headers)
resp.encoding= 'utf-8'
s = resp.text.replace(u' &nbsp;','')#将 &nbsp;替换为空字符，解决空格问题
print(s)
```

```python
obj = re.compile(r'</span></p></td><td style="width:9.36%; border-right-style:solid; border-right-width:0.75pt; border-bottom-style:solid; border-bottom-width:0.75pt; padding-right:5.03pt; padding-left:5.4pt; vertical-align:middle"><p style="margin-top:0pt; margin-bottom:0pt; text-align:center; font-size:10pt"><span style="font-family: 微软雅黑, &quot;Microsoft YaHei&quot;;">(?P<答辩人>.*?)</span></p>'
                 r'</td><td style="width:15.64%; border-right-style:solid; border-right-width:0.75pt; border-bottom-style:solid; border-bottom-width:0.75pt; padding-right:5.03pt; padding-left:5.4pt; vertical-align:middle"><p style="margin-top:0pt; margin-bottom:0pt; text-align:center; font-size:10pt"><span style="font-family: 微软雅黑, &quot;Microsoft YaHei&quot;;">(?P<答辩人专业>.*?)</span></p>'
                 r'</td><td style="width:9.8%; border-right-style:solid; border-right-width:0.75pt; border-bottom-style:solid; border-bottom-width:0.75pt; padding-right:5.03pt; padding-left:5.4pt; vertical-align:middle"><p style="margin-top:0pt; margin-bottom:0pt; text-align:center; font-size:10pt"><span style="font-family: 微软雅黑, &quot;Microsoft YaHei&quot;;">(?P<指导教师>.*?)</span></p>'
                 r'</td><td style="width:59.94%; border-right-style:solid; border-right-width:0.75pt; border-bottom-style:solid; border-bottom-width:0.75pt; padding-right:5.03pt; padding-left:5.4pt; vertical-align:middle"><p style="margin-top:0pt; margin-bottom:0pt; font-size:10pt"><span style="font-family: 微软雅黑, &quot;Microsoft YaHei&quot;;">(?P<论文题目>.*?)</span></p>',re.S)
#注意英文'
result = obj.finditer(s)

f = open("北京师范大学地理科学学部2021届本科毕业论文答辩安排.csv",mode="w",newline='')#newline=''解决空行问题
csvwriter =csv.writer(f)

for it in result:
#     print(it.group("答辩人"))
#     print(it.group("答辩人专业"))
#     print(it.group("论文题目"))
#     print(it.group("指导教师"))
    dic = it.groupdict()
    dic['指导教师'] = dic['指导教师'].strip()
    csvwriter.writerow(dic.values())
    
f.close()
print("over!")
```
